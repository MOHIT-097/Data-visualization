```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))

# Create the bubble chart
plt.scatter(df['GDP'], df['Population'],
            s=df['Population'] * 10,  # Scale bubble size
            alpha=0.5,
            edgecolors='w',
            linewidth=2)

# Add annotations for each country
for i in range(len(df)):
    plt.annotate(df['Country'][i],
                 (df['GDP'][i], df['Population'][i]),
                 fontsize=9, ha='right')

# Set titles and labels
plt.title('Bubble Chart of Country Population vs GDP')
plt.xlabel('GDP (in billions)')
plt.ylabel('Population (in millions)')
plt.grid(True)

# Show the plot
plt.show()
```

```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook

# Set up output to display in the notebook
output_notebook()

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)
# Calculate y values as the sine of x
y = np.sin(x)

# Create a new plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)

# Add a line renderer for the sine wave
p.line(x, y, legend_label="sin(x)", line_width=2, line_color='blue')

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource

# Set up the notebook output
output_notebook()

# Define the data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a bar chart
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruits', y_axis_label='Counts', width=800, height=400)

# Add bars to the plot
p.vbar(x='fruits', top='counts', source=source, width=0.9, color='blue')

# Show the plot
show(p)
```

```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook

# Set up the notebook output
output_notebook()

# Generate random data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency', width=800, height=400)

# Add a quad renderer for the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='blue', line_color='black', alpha=0.7)

# Show the plot
show(p)
```

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, LinearColorMapper, ColumnDataSource

# Set up the notebook output
output_notebook()

# Generate random data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)

# Flatten the data and create arrays for Bokeh
data_flat = data_heatmap.flatten()
x_flat = np.repeat(x, len(y))
y_flat = np.tile(y, len(x))

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x_flat, y=y_flat, value=data_flat))

# Create a Bokeh figure
p = figure(title="Heatmap of Random Data",
           x_axis_label='X-axis',
           y_axis_label='Y-axis',
           x_range=(0, 1),
           y_range=(0, 1),
           width=600,
           height=600)

# Create a color mapper
mapper = LinearColorMapper(palette="Viridis256", low=data_flat.min(), high=data_flat.max())

# Add rectangles for the heatmap
p.rect(x='x', y='y', width=0.1, height=0.1,
        source=source,
        line_color=None,
        fill_color=linear_cmap('value', palette="Viridis256", low=data_flat.min(), high=data_flat.max()))

# Add color bar
color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
